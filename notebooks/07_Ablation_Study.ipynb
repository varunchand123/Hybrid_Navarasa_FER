# 07_Ablation_Study

## Purpose

This notebook presents the Stage-1 ablation study reported in the Navarasa facial expression recognition manuscript. It compares the individual ViT and Swin branches with the proposed Hybrid (ViT + Swin) architecture using the final experimental results reported in the paper.


```
from google.colab import drive
drive.mount('/content/drive')
```

```
import os
import pandas as pd
from IPython.display import display

BASE_DIR = "/content/drive/MyDrive/NAVARASA_PROJECT_DRIVE1"
RESULTS_DIR = os.path.join(BASE_DIR, "results")
os.makedirs(RESULTS_DIR, exist_ok=True)
```

```
ablation_table = pd.DataFrame({

    "Model Variant":[
        "ViT-only",
        "Swin-only",
        "Proposed Hybrid"
    ],

    "ViT Branch":[
        "✓",
        "–",
        "✓"
    ],

    "Swin Branch":[
        "–",
        "✓",
        "✓"
    ],

    "Fusion Strategy":[
        "–",
        "–",
        "Feature Concatenation + MLP"
    ],

    "Accuracy (%)":[
        76.46,
        67.86,
        78.09
    ],

    "Macro F1-score (%)":[
        73.90,
        55.88,
        75.53
    ],

    "Total Parameters (M)":[
        85.806,
        86.752,
        173.481
    ],

    "Computational Cost (GFLOPs)":[
        16.848,
        15.169,
        32.016
    ]
})

ablation_table["Accuracy (%)"] = ablation_table["Accuracy (%)"].map(lambda x:f"{x:.2f}")
ablation_table["Macro F1-score (%)"] = ablation_table["Macro F1-score (%)"].map(lambda x:f"{x:.2f}")
ablation_table["Total Parameters (M)"] = ablation_table["Total Parameters (M)"].map(lambda x:f"{x:.3f}")
ablation_table["Computational Cost (GFLOPs)"] = ablation_table["Computational Cost (GFLOPs)"].map(lambda x:f"{x:.3f}")

display(ablation_table)
```

```
csv_path = os.path.join(RESULTS_DIR, "Table_Stage1_Ablation_Study.csv")
xlsx_path = os.path.join(RESULTS_DIR, "Table_Stage1_Ablation_Study.xlsx")

ablation_table.to_csv(csv_path, index=False)
ablation_table.to_excel(xlsx_path, index=False)

print("CSV saved to :", csv_path)
print("Excel saved to :", xlsx_path)
```
